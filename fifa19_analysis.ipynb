```python
# Keep the columns I find interesting, delete rest
choses_columns = ['Name',
    'Age',
    'Nationality',
    'Overall',
    'Potential',
    'Special',
    'Acceleration',
    'Aggression',
    'Agility',
    'Balance',
    'BallControl',
    'Body Type',
    'Composure',
    'Crossing',
    'Curve',
    'Club',
    'Dribbling',
    'FKAccuracy',
    'Finishing',
    'GKDiving',
    'GKHandling',
    'GKKicking',
    'GKPositioning',
    'GKReflexes',
    'HeadingAccuracy',
    'Interceptions',
    'International Reputation',
    'Jersey Number',
    'Jumping',
    'Joined',
    'LongPassing',
    'LongShots',
    'Marking',
    'Penalties',
    'Position',
    'Positioning',
    'Preferred Foot',
    'Reactions',
    'ShortPassing',
    'ShotPower',
    'Skill Moves',
    'SlidingTackle',
    'SprintSpeed',
    'Stamina',
    'StandingTackle',
    'Strength',
    'Value',
    'Vision',
    'Volleys',
    'Wage',
    'Weak Foot',
    'Work Rate']
```

```python
# Create dataset with chosen columns
df = pd.DataFrame(df_fifa19, columns = choses_columns)
df.sample(5) # 5 random samples
```

```python
# *** ANALYSIS *** 

# Check how the different columns correlates to eachother
# Example longshots vs volleys (expects a close correlation)
df['LongShots'].corr(df['Volleys']) # Gives 0.868, quite high as excpected
df.corr() # Gives correlation between ll columns vs all columns

# Making a correlation heatmap for all columns vs all colums
plt.rcParams['figure.figsize'] = (20, 13)

# Selects the columns I want to see the correlation heatmap for
new_df = df[['Age', 'Overall', 'Potential', 'Value', 'Wage',
                'Acceleration', 'Aggression', 'Agility', 'Balance', 'BallControl', 
                'Body Type','Composure', 'Crossing','Dribbling', 'FKAccuracy', 'Finishing', 
                'HeadingAccuracy', 'Interceptions','International Reputation',
                'Joined', 'Jumping', 'LongPassing', 'LongShots',
                'Marking', 'Penalties', 'Position', 'Positioning',
                'ShortPassing', 'ShotPower', 'Skill Moves', 'SlidingTackle',
                'SprintSpeed', 'Stamina', 'StandingTackle', 'Strength', 'Vision',
                'Volleys']]
hm = sns.heatmap(new_df.corr(), annot=True, linewidth=.5, cmap='Blues')
hm.set_title(label='Heatmap of correlations', fontsize=20)
```

```python
# Scatter plot shows correlation between some feature and some other feature/s
def make_scatter(df, feature, other):
    #feats = ('Agility', 'Balance', 'Dribbling', 'SprintSpeed')
    for index, feat in enumerate(other):
        plt.subplot(len(other)/4+1, 4, index+1)
        ax = sns.regplot(x = feature, y = feat, data = df)

plt.figure(figsize = (20, 20))
plt.subplots_adjust(hspace = 0.4)
make_scatter(df, 'Acceleration', ['Agility', 'Balance', 'Dribbling', 'SprintSpeed'])
```

```python
# Interested in correlations greater than 0.7
interesting_corrs = {}
for x in new_df.columns:
    for y in new_df.columns:
        if new_df[x].dtype == np.float64 and new_df[y].dtype == np.float64:
            corr = new_df[x].corr(new_df[y])
            if corr >= 0.7:
                interesting_corrs[(x, y)] = corr

interesting_corrs

            
```

```python
# Histogram: distribution of playerage 
sns.set(style='dark', palette='colorblind', color_codes=True)
x = df.Age
plt.figure(figsize=(12, 8))
ax = sns.distplot(x, bins=58, kde=False, color='g')
ax.set_xlabel(xlabel='Player\'s age', fontsize=16)
ax.set_ylabel(ylabel='Number of players', fontsize=16)
ax.set_title(label="Histogram of players age", fontsize=20)
plt.show()
```

```python
# Five eldest players
eldest = df.sort_values('Age', ascending=False)
eldest = eldest[['Name', 'Nationality', 'Age']]
eldest.set_index('Name', inplace=True)
eldest.head(3)
```

```python
# Compare six clubs in relation to age
some_clubs = ['Juventus', 'Real Madrid', 'Paris Saint-Germain', 'FC Barcelona', 'Legia Warszawa', 'Manchester United']
df_club = df.loc[df['Club'].isin(some_clubs) & df['Age']]
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
ax = sns.violinplot(x = 'Club', y = 'Age', data=df_club)
ax.set_title(label='Distribution of age in some clubs', fontsize=20)
```

```python
# Longest membership in the club
now = datetime.datetime.now()
df['Joined']
df['JoinYear'] = df.Joined.dropna().map(lambda x: x.split(',')[1].split(' ')[1])
df['YearsInClub'] = df.JoinYear.dropna().map(lambda x: now.year - int(x)).astype(int)

membership = df[['Name', 'Club', 'YearsInClub']].sort_values(by='YearsInClub', ascending=False).dropna().head()
membership.set_index('Name', inplace=True)
membership

```
