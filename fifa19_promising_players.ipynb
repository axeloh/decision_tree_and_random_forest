```python
# * * * FIFA 19 Analysis - Most promising players acording to overall and evolution * * * 
import pandas as pd
players = pd.read_csv('../datasets/fifa19_complete_player_dataset.csv')
players.head()
```

```python
# Creating a new column named 'Difference' to identify which players has more points to evolute
players['Difference'] = players['Potential'] - players['Overall']
```

```python
# Function to classify players who are stable (reached apex), or has small, medium or big evolution
def evolution(d):
    if d == 0:
        return 'Stable'
    elif d>=1 and d <=5:
        return 'Small'
    elif d >=6 and d <=10:
        return "Medium"
    elif d>11:
        return "Big"
```

```python
# Making another column for the evolution
players['Evolution'] = players['Difference'].apply(evolution)
```

```python
players.head()
```

```python
# Filter out players with potential > 80 and that will have a big evolution (11 or more overall points)
x = players.loc[(players['Evolution'] == 'Big') & (players['Potential'] > 80)]
```

```python
# And now in decrescent order
x.sort_values(by="Potential", ascending=False)
```
